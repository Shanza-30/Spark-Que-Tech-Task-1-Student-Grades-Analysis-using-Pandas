**Step 3: Calculate Average Score**

```python
df['Average'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)
df.head()
```

**Step 4: Find Top Performer**

```python
top_student = df.loc[df['Average'].idxmax()]
print("Top Performer:\n", top_student[['gender','Average']])
```

**Step 5: Subject-wise Max & Min Scores**

```python
print("Max Scores:\n", df[['math score','reading score','writing score']].max())
print("Min Scores:\n", df[['math score','reading score','writing score']].min())
```

**Step 6: Subject-wise Average**

```python
print("Average Scores per Subject:\n", df[['math score','reading score','writing score']].mean())
```

**Step 7: Histogram of Average Scores**

```python
plt.hist(df['Average'], bins=5, color='skyblue', edgecolor='black')
plt.title("Distribution of Average Scores")
plt.xlabel("Average Score")
plt.ylabel("Number of Students")
plt.show()
```

**Step 8: Bar Chart of Top 5 Students**

```python
top5 = df.nlargest(5, 'Average')
plt.bar(top5.index, top5['Average'], color='orange')
plt.title("Top 5 Students by Average Score")
plt.xlabel("Student Index")
plt.ylabel("Average Score")
plt.show()
```

**Step 9: Optional Grade Categorization**

```python
def grade(avg):
    if avg>=90: return 'A'
    elif avg>=80: return 'B'
    elif avg>=70: return 'C'
    else: return 'D'

df['Grade'] = df['Average'].apply(grade)
df.head()
```
